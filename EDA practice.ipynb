```python
lst=[]
for age in data['age']:
    if age <=40:
        lst.append("Upto 40")
    elif age>=41 and age<=50:
        lst.append("41-50")
    elif age>=51 and age<=60:
        lst.append("51-60")
    elif age>=61 and age<=70:
        lst.append("61-70")
    else:
        lst.append("Above 71")
data['Age Range']= lst
```

```python
## which age range has highest fbs & cholestrol

sns.barplot(x="Age Range", y="fbs", data=data)
```
