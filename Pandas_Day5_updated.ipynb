## Data Cleaning & Missing Value Handling 

```python
# If the number of entries is N, you have less than 5-10% missing values-- we can eliminate those rows

hotel_data=hotel_data[hotel_data['children'].notna()]
hotel_data.reset_index(inplace=True, drop=True)

hotel_data=hotel_data[hotel_data['country'].notna()]
hotel_data.reset_index(inplace=True, drop=True)

# If the number of entries is N, you have 30-50% missing values-- we can replace the rows with mean, median, mode
hotel_data['agent']= hotel_data['agent'].fillna(hotel_data['agent'].mode()[0])

#If the number of entries is N, you have above 70% missing values-- Use some intuition and eliminate the column

hotel_data=hotel_data.drop(['company'],axis=1)
```

```python
hotel_data.isna().sum()
```

```python
#Drop duplicates again
hotel_data=hotel_data.drop_duplicates()
hotel_data.reset_index(inplace=True, drop=True)
hotel_data
```

```python
#convert datatypes if required
hotel_data['children'] = hotel_data['children'].astype('int64')
hotel_data['agent'] = hotel_data['agent'].astype('int64')
hotel_data['reservation_status_date']= pd.to_datetime(hotel_data['reservation_status_date'])
```

```python
d = {'January':1, 'February':2, 'March':3, 'April':4, 'May':5, 'June':6, 'July':7,'August':8, 'September':9, 'October':10,'November':11,'December':12  }
```

```python
hotel_data['arrival_date_month'] = hotel_data['arrival_date_month'].map(d)
```

```python
# Removing arrival_date_week_number column to avoid redundancy
hotel_data=hotel_data.drop(['arrival_date_week_number'],axis=1)
```
